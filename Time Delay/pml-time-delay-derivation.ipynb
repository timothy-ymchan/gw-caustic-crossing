# Derivation of time delay for point mass lens (with source term)

Here is a reminder of the derivation of point mass lensing. Recall the schwarzschild metric:
\begin{align*}
    d^2s = - (1-r_s/r) dt^2 + (1-r_s/r)^{-1} d^2 r +r^2 (d^2 \theta + \sin^2 \theta d^2 \phi)
\end{align*}
where $r_s=2GM/c^2$ is the Schwarzschild radius. Now we want to derive the geodesic equation for gravitational waves. Without lost of generality we may assume $\theta=\pi/2$, since the gravitational waves travels in null geodesic, we have:
\begin{align*}
    - (1-r_s/r) \left(\frac{dt}{d\lambda}\right)^2 + (1-r_s/r)^{-1} \left(\frac{dr}{d\lambda}\right)^2 +r^2 \left(\frac{d\phi}{d\lambda}\right)^2 = 0
\end{align*}
Now we note that $dt/d\lambda = (1-r_s/r)^{-1} E$, $d\phi/d\lambda = L/r^2$, where $L$ is the initial angular momentum. Substituting we obtain:
\begin{align*}
    - (1-r_s/r)^{-1} E^2+ (1-r_s/r)^{-1} \left(\frac{dr}{d\lambda}\right)^2 +\frac{L^2}{r^2} = 0
\end{align*}
Therefore, we can write the equation of motion as:
\begin{align*}
    \left(\frac{dr}{d\lambda}\right)^2 &= E^2 - \left(1-\frac{r_s}{r}\right) \frac{L^2}{r^2}\\
    \frac{dr}{d\lambda}&=\left[E^2 - \left(1-\frac{r_s}{r}\right) \frac{L^2}{r^2}\right]^{1/2}
\end{align*}
We can thus define the effective potential:
\begin{align*}
    V(r) =  \left(1-\frac{r_s}{r}\right) \frac{L^2}{r^2}
\end{align*}
We can also derive the radius of closes arrival $r_0$, which is when $dr/d\lambda = 0$:
\begin{align*}
    E^2 - \left(1-\frac{r_s}{r_0}\right) \frac{L^2}{r_0^2} = 0
\end{align*}
Here we note that if $\xi$ is the impact parameter, then $L=p\xi=E\xi$ in $c=1$ unit. Note that $[\xi] = \text{length}$. This means the $f_0$ can be solved given $\xi$:
\begin{align*}
    \frac{r_0^2}{\xi^2} = 1 - \frac{r_s}{r_0}
\end{align*}
From now on, we shall just treat $r_0(\xi)$ as a function of impact parameter.
In any case, we see that the time delay is given by:
\begin{align*}
    \frac{dt}{dr} &= (1-r_s/r)^{-1} E\left[E^2 - \left(1-\frac{r_s}{r}\right) \frac{L^2}{r^2}\right]^{-1/2}\\
    &= (1-r_s/r)^{-1} \left[1 - \left(1-\frac{r_s}{r}\right) \frac{\xi^2}{r^2}\right]^{-1/2}\\
    &= (1-r_s/r)^{-1} \left[1 - \left(1-\frac{r_s}{r}\right)\left(1-\frac{r_s}{r_0}\right)^{-1} \frac{r_0^2}{r^2}\right]^{-1/2}
\end{align*}
In the last step we have use relation between $r_0$ and $\xi$ stated above.
Hence, the time delay, give the impact parameter $\xi$ is given as:
\begin{align*}
    t = \int_{r_0}^{R} (1-r_s/r)^{-1} \left[1 - \left(1-\frac{r_s}{r}\right)\left(1-\frac{r_s}{r_0}\right)^{-1} \frac{r_0^2}{r^2}\right]^{-1/2} dr 
\end{align*}
We note that in general $r_0 \ll R$ and $r_s\ll r_0$ so to get a good approximation we can let $u=r_0/r$, $h=r_s/r_0$ and $w=r_0/R$, doing the substitution will lead to:
\begin{align*}
    t = r_0 \int_{w}^{1} u^{-2} (1-hu)^{-1} \left[1-\frac{1-hu}{1-h} u^2 \right]^{-1/2} du
\end{align*}
Note that the 0th order time is simply $R$, so we can also factor it out:
\begin{align*}
    t/R = w\int_{w}^{1} u^{-2} (1-hu)^{-1} \left[1-\frac{1-hu}{1-h} u^2 \right]^{-1/2} du
\end{align*}
Now we have two small quantities $w,h$ we wish to know which of them are larger, to do that, we note that $r_0$ should be of order of the Einstein radius $r_0 \sim \theta_E R \sim \sqrt{r_s R}$. Therefore, we have $w=r_0/R \sim \sqrt{r_s/R}$ and $h\sim r_s/\sqrt{r_s R}\sim \sqrt{r_s/R}$ which is of same order. Now we can expand the integral, first with respect to $h$:
\begin{align*}
    t/R &= w \int_{w}^{1} \frac{1}{u^2 \sqrt{1-u^2}} + \frac{(2+3u)}{2u(1+u)\sqrt{1-u^2}} h +\mathcal{O}(h^2) du\\
    &= \sqrt{1-w^2} + \frac{1}{2} h w\left[\sqrt{\frac{1-w}{1+w}}+2 \ln \left(\frac{1+\sqrt{1-w^2}}{w}\right)\right] + \mathcal{O}(\cdots)^3
\end{align*}
Note that the expansion is asymptotic in the sense that the logarithm diverges when $w$ goes to 0. Now expand in $w$ up to second order:
\begin{align*}
    t/R = 1 - \frac{w^2}{2} + \frac{hw}{2} \left(1 - 2 \ln \frac{w}{2}\right) + \cdots
\end{align*}
This can be rewrite as:
\begin{align*}
    t = R - \frac{R}{2} \left(r_0/R\right)^2 + \frac{r_s}{2} \left(1-2\ln r_0/2R\right) 
\end{align*}
%Now, observing that $w\int_w^1 (\cdots) du = w\int_0^1 (\cdots) du - w \int_0^w (\cdots) du$, the second term is of order $\mathcal{O}(w^2)$. Thus:
\begin{align*}
    t/R &= w \int_{0}^{1} \frac{1}{u^2 \sqrt{1-u^2}} du \\&+wh\int_0^1\frac{(2+3u)}{2u(1+u)\sqrt{1-u^2}}  du  -w^2 \int_0^1 \frac{1}{u^2 \sqrt{1-u^2}} du + \mathcal{O}(\epsilon^3)
\end{align*}

The time delay can be broken into two parts, that from source to lens and from lens to observer. If we do the calculation (See Creeton Sec 3.4.2 for example), we will obtain the following expression:

\begin{align*}
    t = \frac{d_l d_s}{2 d_{ls}} (\theta-\beta)^2 + r_s - r_s \ln\left(\frac{\theta^2}{4} \frac{d_l}{d_{ls}}\right) = \frac{d_l d_s}{2 d_{ls}} (\theta-\beta)^2 -  2 r_s \ln|\theta| + r_s - r_s\ln\left(\frac{d_l}{4 d_{ls}}\right)
\end{align*}
Note that the last few terms are roughly constant, even if the lens is moving, because the lens and source usually moves in a length scale much smaller than that of their separation.
In conclusion, the time delay is:
\begin{align*}
    t = \underbrace{\frac{d_l d_s}{2 d_{ls}} (\theta-\beta)^2}_\text{Path diff.} \quad \underbrace{-4M_L\ln|\theta|}_{\phi(\vec \theta)} \quad + \underbrace{d_s \left(1+\frac{1}{2}\beta^2\right)}_{\psi(\vec \beta)}\quad \underbrace{- 2M_L \left[1+\ln\left(\frac{d_l}{4 d_s}\right)\right]}_\text{const.}
\end{align*}
where $\theta,\beta$ are the image and source angle respectively and $M_L$ is the lens mass.

Note that $\psi(\vec \beta)$ is the time delay that would have experienced by the ray if it travelled on a straight line.
